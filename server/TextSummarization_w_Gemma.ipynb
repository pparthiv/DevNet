```python
import os
import json
from langchain import PromptTemplate
from langchain_community.llms import HuggingFaceEndpoint
from langchain import LLMChain
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python


template = """You are tool capable of providing expert, concise, and critical summaries based on a developer's skills, and their certifications. Your task is to generate a new summary as a paragraph that effectively represents the developer's profile containing important information and the fields they can work in, while disregarding any other information not related to their development profile, and avoid writing their skills, instead show how their certifications tie in their working fields.
Skills: {skills},
Certifications: {certifications}
Answer only as the summary paragraph
"""
prompt = PromptTemplate(
    template=template,
    input_variables=['about', 'skills', 'certifications']
)
```

```python
gemma_7b = HuggingFaceEndpoint(repo_id='google/gemma-7b-it', temperature=0.3, token=os.environ["HUGGINGFACEHUB_API_TOKEN"])
```

```python
candidate = open("./text.json")
```

```python
candidate_data = json.load(candidate)
```

```python
about = candidate_data["Summary"]
skills = candidate_data["Skills"]
certs = candidate_data["Certifications"]
```

```python
llm_chain = LLMChain(
    prompt=prompt,
    llm=gemma_7b
)

# ask the question about Kaggle
res = llm_chain.run({"skills":skills, "certifications":certs})
```
